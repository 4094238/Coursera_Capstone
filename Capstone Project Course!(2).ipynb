### Part 1 - Postal Code / Borough / Neighborhood Table

```python
#getting the data

import urllib.request, urllib.parse, urllib.error
import json
url = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
fhand = urllib.request.urlopen(url)
print(fhand)
```

```python
#turning the http response into a pandas dataframe

df = pd.read_html(url)
df = df[0]
print(type(df))
print(df.shape)
df.head(10)
```

```python
#dropping all df["Borough"]=="Not assigned"

df.set_index("Borough", inplace = True)
df.drop("Not assigned", axis = 0, inplace = True)
df.reset_index(inplace = True)
print(df.shape)
df.head(10)
```

### Part 2 - Adding Lat and Long to our table

```python
import geocoder
```

```python
# empty lists

latitude = list() 
longitude = list()

# looping to generate list with lat and long for each postal code

for i in df['Postal Code']:
    g = geocoder.arcgis("{}, Toronto, Ontario".format(i))
    lat = g.json['lat']
    lng = g.json['lng']
    latitude.append(lat)
    longitude.append(lng)

# printing to check everything went well

print("latitude type", type(latitude))
print('longitude type', type(longitude))
print(latitude)
print(longitude)

# adding latitude and lingitude columns to our dataframe

df['Latitude'] = latitude
df['Longitude'] = longitude

# checking everything went well

print(df.shape)
df.head()
```

```python
# function to explore each Neighborhood

import requests
def getNearbyVenues(names, latitudes, longitudes, radius=500):
    
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        print(name)
            
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)
            
        # make the GET request
        results = requests.get(url).json()["response"]['groups'][0]['items']
        
        # return only relevant information for each nearby venue
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Neighborhood', 
                  'Neighborhood Latitude', 
                  'Neighborhood Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venues)
```

```python
#Dataframe for the dataset that includes the explored neighborhoods

toronto_venues = getNearbyVenues(df['Neighbourhood'], df['Latitude'], df['Longitude'], radius = 500)
```

```python
# getting dummies

toronto_onehot = pd.get_dummies(toronto_venues[['Venue Category']], prefix = "", prefix_sep = "")
toronto_onehot['Neighborhood'] = toronto_venues['Neighborhood']
fixed_columns = [toronto_onehot.columns[-1]] + list(toronto_onehot.columns[:-1])
toronto_onehot = toronto_onehot[fixed_columns]
toronto_onehot.head()
```

```python
toronto_onehot.shape
```

```python
#dataframe with all venues for each Neighborhood

toronto_grouped = toronto_onehot.groupby('Neighborhood').mean().reset_index()
toronto_grouped
```

```python
toronto_grouped.shape
```

```python
# auxiliary function to get the most common venues

def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]
```

```python
# getting the dataframe with top 10 venues per neighborhood

num_top_venues = 10
indicators = ['st', 'nd', 'rd']

columns = ['Neighborhood']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))

neighborhoods_venues_sorted = pd.DataFrame(columns=columns)
neighborhoods_venues_sorted['Neighborhood'] = toronto_grouped['Neighborhood']

for ind in np.arange(toronto_grouped.shape[0]):
    neighborhoods_venues_sorted.iloc[ind, 1:] = return_most_common_venues(toronto_grouped.iloc[ind, :], num_top_venues)

print(neighborhoods_venues_sorted.shape)
neighborhoods_venues_sorted.head()
```

```python
# clustering: fitting KMeans

kclusters = 6
toronto_grouped_clustering = toronto_grouped.drop('Neighborhood', 1)
kmeans = KMeans(n_clusters = kclusters, random_state = 0).fit(toronto_grouped_clustering)
kmeans.labels_
```

```python
toronto_grouped_clustering.head()
```

```python
neighborhoods_venues_sorted.insert(0, "Cluster Labels", kmeans.labels_)
neighborhoods_venues_sorted
```

```python
# merging to add latitude and longitude to our dataframe

toronto_merged = neighborhoods_venues_sorted
toronto_merged = toronto_merged.join(df.set_index('Neighborhood'),\
                                     on = 'Neighborhood', how = 'left')
toronto_merged
```

```python
toronto_merged['Cluster Labels'].values
```

```python
# showing the clusters in the map

map_clusters = folium.Map(location = [toronto_lat, toronto_lng],tiles='OpenStreetMap', zoom_start = 11)

x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

markers_colors = []
for lat, lon, poi, cluster in zip(toronto_merged['Latitude'], toronto_merged['Longitude'], toronto_merged['Neighborhood'], toronto_merged['Cluster Labels']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
map_clusters
```
