#### Postal Code / Borough / Neighborhood Table

```python
#getting the data

import urllib.request, urllib.parse, urllib.error
import json
url = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
fhand = urllib.request.urlopen(url)
print(fhand)
```

```python
#checking the http response

for line in fhand:
    print(line)
```

```python
#turning the http response into a pandas dataframe

df = pd.read_html(url)
df = df[0]
print(type(df))
print(df.shape)
df.head(10)
```

```python
#dropping all df["Borough"]=="Not assigned"

df.set_index("Borough", inplace = True)
df.drop("Not assigned", axis = 0, inplace = True)
df.reset_index(inplace = True)
print(df.shape)
df.head(10)
```

```python
# checking there are no any df["Neighbourhood"]=="Not assigned"
for i in df["Neighbourhood"]:
    try:
        df["Neighbourhood"][i] == "Not assigned"
        print("replace", i, "element")
    except:
        pass
print("finish")
```
